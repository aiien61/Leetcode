```python
from typing import List, Optional

class BST:
    def __init__(self, nums: List[int]):
        self.nums = nums

    def search(self, target_val: int) -> Optional[int]:

        index_start, index_end = 0, len(self.nums) - 1
        index_root = (index_start + index_end) // 2
        
        if not self.nums[index_root]:
            return None
        
        return self._search(target_val, index_start, index_end)

    def _search(self, target_val: int, index_start: int, index_end: int) -> Optional[int]:
        if index_start > index_end:
            return None
        
        index_mid = (index_start + index_end) // 2

        if target_val == self.nums[index_mid]:
            return target_val
        
        elif target_val < self.nums[index_mid]:
            return self._search(target_val, index_start, index_mid-1)
        
        elif target_val > self.nums[index_mid]:
            return self._search(target_val, index_mid+1, index_end)
        
        return None
    
    def search_leftmost(self, target_val: int) -> Optional[int]:

        index_start, index_end = 0, len(self.nums) - 1
        index_root = (index_start + index_end) // 2
        
        if not self.nums[index_root]:
            return None
        
        return self._search_leftmost(target_val, index_start, index_end)

    def _search_leftmost(self, target_val: int, index_start: int, index_end: int) -> Optional[int]:
        if index_start > index_end:
            return None
        
        index_mid = (index_start + index_end) // 2

        if target_val == self.nums[index_mid]:
            result = self.nums[index_mid]
            result_left = self._search_leftmost(target_val, index_start, index_mid-1)
            if result_left:
                result = result_left
                print('(found more-left result)', end="")
            return result
        
        elif target_val < self.nums[index_mid]:
            return self._search_leftmost(target_val, index_start, index_mid-1)
        
        elif target_val > self.nums[index_mid]:
            return self._search_leftmost(target_val, index_mid+1, index_end)
        
        return None
    
    def search_rightmost(self, target_val: int) -> Optional[int]:

        index_start, index_end = 0, len(self.nums) - 1
        index_root = (index_start + index_end) // 2
        
        if not self.nums[index_root]:
            return None
        
        return self._search_rightmost(target_val, index_start, index_end)

    def _search_rightmost(self, target_val: int, index_start: int, index_end: int) -> Optional[int]:
        if index_start > index_end:
            return None
        
        index_mid = (index_start + index_end) // 2

        if target_val == self.nums[index_mid]:
            result = self.nums[index_mid]
            result_right = self._search_rightmost(target_val, index_mid+1, index_end)
            if result_right:
                result = result_right
                print('(found more-right result)', end="")
            return result
        
        elif target_val < self.nums[index_mid]:
            return self._search_rightmost(target_val, index_start, index_mid-1)
        
        elif target_val > self.nums[index_mid]:
            return self._search_rightmost(target_val, index_mid+1, index_end)
        
        return None
```

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        if not nums: return [-1, -1]
        
        return self._searchRange(nums=nums, target=target, index_low=0, index_high=len(nums)-1)

    def _searchRange(self, nums: list, target: int, index_low: int, index_high: int) -> List[int]:
        result = [-1, -1]
        
        if index_low > index_high:
            return result
        
        index_mid = (index_low + index_high) // 2

        if target == nums[index_mid]:
            result[0], result[1] = index_mid, index_mid
            
            if index_low < index_mid:
                result_tmp_left = self._searchRange(nums, target, index_low, index_mid-1)
                if result_tmp_left[0] != -1:
                    result[0] = result_tmp_left[0]
            
            if index_mid < index_high:
                result_tmp_right = self._searchRange(nums, target, index_mid+1, index_high)
                if result_tmp_right[1] != -1:
                    result[1] = result_tmp_right[1]
            
            return result
        
        elif target < nums[index_mid]:
            result = self._searchRange(nums, target, index_low, index_mid-1)
        
        elif target > nums[index_mid]:
            result = self._searchRange(nums, target, index_mid+1, index_high)
        
        return result
```

```python
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        if not piles: return 0
        
        if h == len(piles): return max(piles)    

        max_banana_number = max(piles)
        return self._minEatingSpeed(piles, h, 1, max_banana_number)
    
    def _minEatingSpeed(self, piles: List[int], h: int, k_low: int, k_high: int) -> int:
        if k_low > k_high: return -1
        
        k_result = -1
        k_result_tmp = -1
        
        k_mid = (k_low + k_high) // 2
        
        total_hours = self.calculate_total_hours(piles, k_mid)

        if total_hours <= h:
            k_result = k_mid
            if k_low < k_mid:
                k_result_tmp = self._minEatingSpeed(piles, h, k_low, k_mid - 1)
                if k_result_tmp != -1:
                    k_result = k_result_tmp
        elif total_hours > h:
            if k_mid < k_high:
                k_result = self._minEatingSpeed(piles, h, k_mid + 1, k_high)
        
        return k_result
    
    def calculate_total_hours(self, piles, k):
        i = 0
        h = 0
        while i < len(piles):
            bananas = piles[i]
            if bananas % k == 0:
                h += bananas // k
            else:
                h += (bananas // k + 1)
            i += 1
        
        return h
        
```

```python
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:
        if not piles: return 0
        
        if h == len(piles): return max(piles)
        
        speed_low, speed_high = 1, max(piles)
        
        # find speed rate as slow as possible but must enough for finishing all piles in h hours
        while speed_low < speed_high:
            speed_mid = (speed_low + speed_high) // 2
            
            total_hours = 0
            for pile in piles:
                if pile % speed_mid == 0:
                    total_hours += (pile // speed_mid)
                else:
                    total_hours += (pile//speed_mid + 1)
            
            if total_hours == h:
                # total hours are exactly equal to h hours, so it is perfect speed rate
                speed_low = speed_mid
                break

            elif total_hours < h:
                # speed is too fast that all piles can be eaten up within h hours, so slow down 
                speed_high = speed_mid - 1
                
            else:
                # speed is too slow that not enough for all piles in h hours, so speed up
                speed_low = speed_mid + 1
        
        return speed_low
```
